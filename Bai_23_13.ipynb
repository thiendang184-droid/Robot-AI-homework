```python
import folium
from folium.plugins import HeatMap
import numpy as np

center_hcm = [10.775, 106.690]
m = folium.Map(location=center_hcm, zoom_start=13)

fg_points = folium.FeatureGroup(name="📍 Lớp Điểm: Hệ thống Đại lý")
fg_areas = folium.FeatureGroup(name="🔥 Lớp Vùng: Mật độ Khách hàng (Heatmap)")
fg_routes = folium.FeatureGroup(name="🛣️ Lớp Tuyến: Tuyến Cung ứng Huyết mạch")

dealers = [[10.78, 106.70], [10.76, 106.68], [10.77, 106.67]]
for i, d in enumerate(dealers):
    folium.Marker(d, popup=f"Đại lý cấp {i+1}", icon=folium.Icon(color='red', icon='store', prefix='fa')).add_to(fg_points)

np.random.seed(42)
demand_data = (np.random.normal(size=(100, 2)) * 0.015 + center_hcm).tolist()
HeatMap(demand_data, radius=15, blur=10).add_to(fg_areas)

warehouse = [10.79, 106.71]
folium.Marker(warehouse, popup="Kho Tổng", icon=folium.Icon(color='black', icon='building', prefix='fa')).add_to(m) # Kho tổng luôn hiện

for d in dealers:
    folium.PolyLine([warehouse, d], color='blue', weight=4, dash_array='5', popup="Tuyến cung ứng").add_to(fg_routes)

m.add_child(fg_areas)  # Vùng nằm dưới cùng
m.add_child(fg_routes) # Tuyến nằm giữa
m.add_child(fg_points) # Điểm nằm trên cùng

folium.LayerControl(collapsed=False).add_to(m)

display(m)
```
