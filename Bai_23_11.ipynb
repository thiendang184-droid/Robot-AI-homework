```python
import folium
from folium.plugins import HeatMapWithTime
import numpy as np

m = folium.Map(location=[10.77, 106.68], zoom_start=13)

#Giả lập dữ liệu cho 3 mốc thời gian (Sáng, Trưa, Tối)
np.random.seed(42)

#Sáng: Tập trung ở Quận 1 (văn phòng)
morning_data = (np.random.normal(size=(50, 2)) * 0.01 + [10.775, 106.695]).tolist()

#Trưa: Lan ra các khu ăn uống Quận 3
noon_data = (np.random.normal(size=(50, 2)) * 0.01 + [10.780, 106.680]).tolist()

#Tối: Tập trung ở khu vui chơi Quận 10
evening_data = (np.random.normal(size=(50, 2)) * 0.01 + [10.765, 106.665]).tolist()

data_over_time = [morning_data, noon_data, evening_data]
time_index = ['08:00 AM', '12:00 PM', '08:00 PM']

HeatMapWithTime(data_over_time, index=time_index, radius=20, auto_play=True, max_opacity=0.8).add_to(m)

m
```
