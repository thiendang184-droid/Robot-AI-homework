```python
!pip install osmnx networkx geopy folium
import folium
import osmnx as ox
import networkx as nx

# 1. Khởi tạo vị trí (Khu vực ngã tư Bảy Hiền - thường xuyên kẹt xe)
location = [10.793, 106.654]
m = folium.Map(location=location, zoom_start=15)

# 2. Giả lập một "Vùng Nguy Cơ Cao" (Màu đỏ rực)
folium.Circle(
    location=location,
    radius=200,
    color='red',
    fill=True,
    fill_opacity=0.4,
    popup="VÙNG TẮC NGHẼN NGHIÊM TRỌNG"
).add_to(m)

# 3. Tìm đường thay thế bằng OSMnx
# Lấy graph quanh khu vực
G = ox.graph_from_point(location, dist=1000, network_type='drive')

# Giả sử ta muốn đi từ A sang B nhưng phải tránh cái lõi kẹt xe ở trên
start_point = [10.790, 106.650]
end_point = [10.796, 106.658]

# Tìm nút gần nhất
orig_node = ox.distance.nearest_nodes(G, X=start_point[1], Y=start_point[0])
dest_node = ox.distance.nearest_nodes(G, X=end_point[1], Y=end_point[0])

# Thuật toán AI: Tìm đường ngắn nhất (Dijkstra)
route = nx.shortest_path(G, orig_node, dest_node, weight='length')
route_coords = [(G.nodes[n]['y'], G.nodes[n]['x']) for n in route]

# 4. Vẽ tuyến đường đề xuất (Màu xanh lá) để tránh vùng kẹt
folium.PolyLine(route_coords, color='green', weight=5, opacity=0.8, popup="Tuyến đường thay thế AI đề xuất").add_to(m)
folium.Marker(start_point, icon=folium.Icon(color='blue')).add_to(m)
folium.Marker(end_point, icon=folium.Icon(color='red')).add_to(m)
m
```
