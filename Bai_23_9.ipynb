```python
!pip install scikit-learn folium numpy

import folium
import numpy as np
from sklearn.cluster import KMeans

np.random.seed(42)
center_city = [10.762, 106.665]

X = (np.random.normal(size=(40, 2)) * 0.015 + center_city).tolist()

n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=0).fit(X)

centers = kmeans.cluster_centers_
labels = kmeans.labels_

m = folium.Map(location=center_city, zoom_start=14, tiles='CartoDB positron')

colors = ['red', 'blue', 'green']

for i in range(len(X)):
    folium.CircleMarker(
        location=X[i],
        radius=5,
        color=colors[labels[i]],
        fill=True,
        fill_opacity=0.6,
        popup=f"Khách hàng nhóm {labels[i]+1}"
    ).add_to(m)

for i, center_pos in enumerate(centers):
    folium.Marker(
        location=center_pos,
        popup=f"VỊ TRÍ ĐẶT TRẠM {i+1} (TỐI ƯU)",
        icon=folium.Icon(color='black', icon='star', prefix='fa')
    ).add_to(m)
display(m)
```
