```python
!pip install osmnx
import folium
from geopy.geocoders import Nominatim
import osmnx as ox
import networkx as nx

geolocator = Nominatim(user_agent="ueh", timeout=10)
location_A = geolocator.geocode("Ben Thanh Market")
location_B = geolocator.geocode("Tan Son Nhat Airport")

latlng_A = (location_A.latitude, location_A.longitude)
latlng_B = (location_B.latitude, location_B.longitude)

midpoint = ((latlng_A[0] + latlng_B[0]) / 2, (latlng_A[1] + latlng_B[1]) / 2)
G = ox.graph_from_point(midpoint, dist=5000, network_type='drive')

orig_node = ox.distance.nearest_nodes(G, X=latlng_A[1], Y=latlng_A[0])
dest_node = ox.distance.nearest_nodes(G, X=latlng_B[1], Y=latlng_B[0])


route_dijkstra = nx.shortest_path(G, orig_node, dest_node, weight='length', method='dijkstra')

route_astar = nx.astar_path(G, orig_node, dest_node, weight='length')

def get_coords(route):
    return [(G.nodes[n]['y'], G.nodes[n]['x']) for n in route]

m = folium.Map(location=midpoint, zoom_start=13)

folium.PolyLine(get_coords(route_dijkstra), color='green', weight=6, opacity=0.6, label='Dijkstra').add_to(m)

folium.PolyLine(get_coords(route_astar), color='red', weight=2, opacity=1, label='A*').add_to(m)

folium.Marker(latlng_A, popup="Điểm A (Bến Thành)", icon=folium.Icon(color='green')).add_to(m)
folium.Marker(latlng_B, popup="Điểm B (Sân bay)", icon=folium.Icon(color='blue')).add_to(m)

display(m)
```
