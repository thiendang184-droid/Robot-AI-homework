```python

!pip install geopy folium

from geopy.geocoders import Nominatim
from geopy.distance import geodesic
import folium

geolocator = Nominatim(user_agent="geoapi")
ueh_coords = [10.761235853524624, 106.66833034179186]

m = folium.Map(location=ueh_coords, zoom_start=15)

folium.Marker(
    location=ueh_coords,
    popup="Đại học UEH cơ sở B",
    icon=folium.Icon(color="blue", icon="graduation-cap", prefix='fa')
).add_to(m)

addresses = ["Chợ Bến Thành", "Công viên số 1 Lý Thái Tổ"]

for addr in addresses:
    location = geolocator.geocode(addr)
    if location:
        point = (location.latitude, location.longitude)
        dist = geodesic(ueh_coords, point).km

        folium.Marker(point, popup=f"{addr}: {dist:.2f}km").add_to(m)
        folium.PolyLine([ueh_coords, point], color="blue", weight=2, opacity=0.5).add_to(m)
m
```
