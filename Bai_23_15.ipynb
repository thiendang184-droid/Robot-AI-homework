```python
!pip install osmnx
import folium
from sklearn.cluster import KMeans
import numpy as np
import osmnx as ox

center = [10.765, 106.665]

G = ox.graph_from_point(center, dist=2000, network_type='drive')


np.random.seed(10)
points = (np.random.normal(size=(100, 2)) * 0.015 + center).tolist()


kmeans = KMeans(n_clusters=3, n_init=10).fit(points)
raw_centers = kmeans.cluster_centers_

m = folium.Map(location=center, zoom_start=14)


for p in points:
    folium.CircleMarker(location=p, radius=2, color='blue', opacity=1).add_to(m)

for hub in raw_centers:

    nearest_node = ox.distance.nearest_nodes(G, X=hub[1], Y=hub[0])
    real_pos = [G.nodes[nearest_node]['y'], G.nodes[nearest_node]['x']]

    folium.Marker(real_pos, popup="TRẠM LOGISTICS TỐI ƯU",
                  icon=folium.Icon(color='darkgreen', icon='truck', prefix='fa')).add_to(m)

m
```
