```python
!pip install osmnx networkx folium
import folium
from folium.plugins import TimestampedGeoJson
import osmnx as ox
import networkx as nx
from datetime import datetime, timedelta

center = [10.765, 106.685]
G = ox.graph_from_point(center, dist=1500, network_type='drive')

start_node = list(G.nodes)[5]
end_node = list(G.nodes)[-5]
route = nx.shortest_path(G, start_node, end_node, weight='length')

route_coords = [[G.nodes[n]['x'], G.nodes[n]['y']] for n in route]

features = []
start_time = datetime(2026, 4, 21, 8, 0)

for i, coord in enumerate(route_coords):
    current_time = (start_time + timedelta(minutes=i)).isoformat()
    features.append({
        'type': 'Feature',
        'geometry': {'type': 'Point', 'coordinates': coord},
        'properties': {
            'time': current_time,
            'popup': f"Xe đang tại nút giao {i}",
            'iconstyle': {'fillColor': 'green', 'fillOpacity': 1, 'stroke': 'false', 'radius': 7}
        }
    })

m = folium.Map(location=center, zoom_start=15)

folium.PolyLine([(c[1], c[0]) for c in route_coords], color='gray', weight=2, dash_array='5').add_to(m)

TimestampedGeoJson({'type': 'FeatureCollection', 'features': features},
                   period='PT1M', add_last_point=True, auto_play=True).add_to(m)
m
```
