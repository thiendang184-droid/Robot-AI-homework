```python
import folium

ueh_coords = [10.761235853524624, 106.66833034179186]

m = folium.Map(location=ueh_coords, zoom_start=15)

folium.Marker(
    location=ueh_coords,
    popup="Đại học UEH cơ sở B",
    icon=folium.Icon(color="blue", icon="graduation-cap", prefix='fa')
).add_to(m)

mall_coords = [10.770680187663258, 106.66983696617629]
folium.Marker(
    location=mall_coords,
    popup="mall gần trường",
    icon=folium.Icon(color="red", icon="shopping-cart", prefix='fa')
).add_to(m)

folium.LayerControl().add_to(m)

m.save("bai23_1.html")
m
```
