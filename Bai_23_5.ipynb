```python
import folium

warehouse = [10.76, 106.66]
m = folium.Map(location=warehouse, zoom_start=12)

zones = [
    {"radius": 10000, "color": "red", "label": "Vùng xa (10km)"},
    {"radius": 5000, "color": "orange", "label": "Vùng mở rộng (5km)"},
    {"radius": 3000, "color": "green", "label": "Vùng ưu tiên (3km)"}
]

for zone in zones:
    folium.Circle(
        location=warehouse,
        radius=zone["radius"],
        color=zone["color"],
        fill=True,
        fill_opacity=0.1,
        popup=zone["label"]
    ).add_to(m)

folium.Marker(warehouse, icon=folium.Icon(icon="home", color="black")).add_to(m)
m
```
