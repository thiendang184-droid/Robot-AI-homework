```python
import folium
from folium.plugins import HeatMap
import numpy as np
m = folium.Map(location=[10.775, 106.700], zoom_start=12)

np.random.seed(30)
data = (np.random.normal(size=(100, 2)) * 0.015 + [10.775, 106.700]).tolist()

HeatMap(data, radius=15, blur=20).add_to(m)
m
# - Vùng màu Đỏ/Vàng (Nóng): Nơi tập trung mật độ đơn hàng cao.
#   => Quản trị: Nên tập trung nhân sự giao hàng (shipper) tại đây
# - Vùng màu Xanh/Trắng (Lạnh): Nơi mật độ đơn hàng thưa thớt.
#   => Quản trị: Cần đẩy mạnh marketing, khuyến mãi tại khu vực này để cư dân sử dụng dịch vụ giao hàng
```
