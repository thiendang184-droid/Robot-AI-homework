```python
import osmnx as ox
import folium

coords = (10.7601, 106.6691)
G = ox.graph_from_point(coords, dist=500, network_type='drive')

# 2. Phân tích cơ bản
nodes_count = len(G.nodes)
edges_count = len(G.edges)
avg_length = sum(d['length'] for u, v, d in G.edges(data=True)) / edges_count

nodes, edges = ox.graph_to_gdfs(G)

m = folium.Map(location=coords, zoom_start=17)

folium.GeoJson(edges).add_to(m)

analysis_html = f'''
<div style="position:fixed; bottom:50px; left:50px; width:280px; background:white;
            padding:15px; border:2px solid black; z-index:9999; font-size:14px;">
    <b>PHÂN TÍCH AI ĐÔ THỊ (23.6):</b><br><br>
    - <b>Số nút giao:</b> {nodes_count}<br>
    - <b>Số đoạn đường:</b> {edges_count}<br>
    - <b>Độ dài TB:</b> {avg_length:.2f} m<br><br>
    <i>Dữ liệu OSMnx cung cấp nền tảng để AI tối ưu hóa tình trạng giao thông.</i>
</div>
'''
m.get_root().html.add_child(folium.Element(analysis_html))

m
```
